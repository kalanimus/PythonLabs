```python
import subprocess
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
config_path = "Files/config2.json"
program_path = "Program/bin/Debug/net8.0/Program.exe"

subprocess.run([program_path, config_path], capture_output=True, text=True)
```

```python
data = pd.read_csv('Files/iceCreamCount.csv')

plt.figure(figsize=(10, 6))
sns.barplot(x='Вкус', y='Количество', data=data)

plt.title('Сколько произведено мороженного')
plt.xlabel('Вкус')
plt.ylabel('Количество')

plt.gca().set_yticks(range(0, data['Количество'].max() + 1))  

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
